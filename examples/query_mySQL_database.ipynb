# Query with conditions

```python
# Eventually build a wrapper to a GUI that constructs the equivalent of tables_dict below:

#nicknames = ["cs", "mono", "ndf", "sr510", "sr830"]
pylab = {'control_software':['timestamp','date','sequence']}
ndf = {'ndf':['position']}
cs260 = {'cs260':['wavelength','grating','order_sort_filter','shutter']}
sr510 = {'sr510':['start_time','sample_rate','time_constant','sensitivity','storage_path']}
sr830 = {'sr830':['start_time','sample_rate','time_constant','sensitivity','storage_path']}

#tables_list = [pylab, cs260, ndf, sr510, sr830]
tables_list = [pylab, cs260, ndf]
tables_dict = {}
for itable in tables_list:
    for key, value in itable.items():
        if key in tables_dict:
            tables_dict[key].extend(value)
        else:
            tables_dict[key] = value
```

```python
# Wrapper will also add conditions like so:

conditions = {"where": {"control_software": {"timestamp": "< CURDATE()"}}}
conditions = {"where": {"control_software": {"timestamp": ">= 2021-12-18"}, 
                        "cs260":{"wavelength": "> 0.5"},}}
```

```python
# Query feeds tables and conditions to query_database and returns pandas dataframe
example_dataframe = example_query.query_database(tables_dict, conditions=conditions)
example_dataframe.head()
```

```python
# You can also print out the statement a
example_query_statement = example_query.build_sql_query(tables_dict, conditions=conditions)
print(example_query.sql_query)
```
